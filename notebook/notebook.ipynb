## Imports

```python
import sys
sys.path.append("..")

import pandas as pd
from datacleaner import DataCleaner
```

## Instancia del objeto `cleaner` + dataset de prueba

```python
df = pd.read_csv("ventas.csv")
df.set_index("id_venta", inplace=True)

cleaner = DataCleaner(df)
cleaner.get_summary()
```

## Datos nulos

```python
cleaner.display_nulls()
```

```python
cleaner.clean_nulls(delete=True)
```

## Datos duplicados

```python
cleaner.clean_dups()
```

## Corrección de tipos de datos

```python
cleaner.cast_columns(columns=["cantidad"], to="int64")
cleaner.cast_columns(columns=["precio"], to="float64")
cleaner.cast_columns(columns=["fecha_venta"], to="datetime")
cleaner.cast_columns(columns=["producto"], to="category")
cleaner.cast_columns(columns=["categoria"], to="category")
```

```python
cleaner.df.dtypes
```

## Verificación de outliers

```python
cleaner.display_outliers(columns=["precio", "cantidad"], method="iqr")
```
